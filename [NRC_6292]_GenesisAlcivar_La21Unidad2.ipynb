# Segundo Error - Valores Nulos

El segundo error encontrado dentro del dataset son valores nulos dentro de las columnas City, States y Range. Mediante el uso de la función isnull podemos observar el número exacto de valores faltantes dentro de las columnas.

```python

    '''Mediante el uso del método isnull visualizamos el total de valores nulos existentes en cada una de las columnas '''

df.isnull().sum()
```

# Corrección del Error

Para corregir el error encontrado se hace uso de la función fillna la cual reemplaza los valores faltantes o nulos por valores en cero, en el conjunto de datos existen valores nulos dentro de las columnas City, State y Range tal como se observa. Al implementar esta función podemos ver que donde existían valores NaN los reemplazo por valores en 0, ya que al contar con valores nulos nos puede generar muchos problemas a la hora de realizar un análisis de nuestros datos.

```python

    '''Creamos una funciòn la cual nos ayudarà a reemplazar los valores vacios que existen en las columnas City, State y Range por valores en Cero, ya que al tener valores nulos nos puede generar muchos problemas a la hora de realizar un análisis de nuestros datos.  '''

df[['City', 'Range','State']] = df[['City', 'Range','State']].fillna(value=0)
```

El tercer error presentado fue el tener datos redundantes en las columnas City y State. En la columna City existian datos de la ciudad de San Francisco en mayúsculas y minúsculas, mientras que en la columna State se transforma CA a 1 para tener la columna totalmente numérica.

# Quinto Error - Datos Duplicados

El quinto error presentado es la existencia de datos duplicados dentro de la columna CrimeId. Mediante el uso de la función duplicated podemos verificar el número exacto de datos duplicados dentro de la columna CrimerId.

```python

    '''Mediante el uso del método duplicated verficamos el número de datos duplicados dentro de la columna CrimeId   '''

df['CrimeId'].duplicated().sum()
```

```python

    '''Creamos un filtro de los datos duplicados de la columna CrimeID y los transformamos en una lista.   '''

duplicates = df[df['CrimeId'].duplicated(keep=False)]['CrimeId'].tolist()
```

```python

    '''Visualizamos los datos de los ID duplicados que existen dentro de la columna CrimeId.   '''

duplicates[:10]
```

# Corrección del Error

Para corregir el error encontrado se crea un set de la lista de los valores, donde se le especifíca mediante el for que se debe recorrer el set de duplicados y debe comparar con cada uno de los datos en el Dataframe. Al encontrar un dato igual este debe sumarle 1 al número de CrimeId, de esta forma evitamos que se den valores duplicados dentro del CrimeId que es donde nos presenta el error.

```python

    '''Realizamos una función la cual crea un set de la lista de los datos duplicados existentes en la columna CrimeId   '''

set_duplicates = set(duplicates)
```

```python
'''Realizamos una función mediante el uso de un for el cual debe recorrer el set de duplicados y debe comparar con cada uno de los datos en el Dataframe. Al encontrar un dato igual este debe sumarle 1 al número de CrimeId, de esta forma evitamos que se den valores duplicados dentro del CrimeId que es donde nos presenta el error. '''
    
for r in set_duplicates:
    dup= df[df['CrimeId'] == r]
    df.loc[dup.iloc[1,:].name, 'CrimeId']= df.loc[dup.iloc[1,:].name]['CrimeId'] + 1    
```

Verificamos una vez más el número de duplicados dentro de la columna mediante la función duplicated y como podemos observar ya no presenta ningún dato duplicado.

```python

    '''Visualizamos nuevamente los datos para verificar que no haya ningun registro duplicado dentro de la columna CrimeId '''

df.duplicated().sum()
```

# Sexto Error - Formato de Fechas y Horas

El sexto error presentado como se puede visualizar dento de las columnas OffenseDate y CallDateTime contienen la misma información y por ende el mismo formato de fecha y hora. Por lo cual se realiza el cambio de formato y del tipo del datos de estas dos columnas. 

```python

    '''Mediante el uso del método loc seleccionamos las filas y las columnas de las columnas OffenseDate y CallDateTime para poder visualizar de mejor manera el tipo de formato que manejan en la Fecha del delito y la Fecha y hora de la llamada '''

df.loc[:, ['OffenseDate', 'CallDateTime']]
```

# Corrección del Error

Para corregir el error encontrado creamos una función mediante el uso del método to_datetime. Este tipo de método nos permite cambiar el formato de la hora y fecha de las columnas OffenseDate, CallDateTime y CallTime.

```python

    '''Creamos una función mediante el uso del método to_datetime para modificar el formato de hora y fecha correctamente dentro de la columna OffenseDate(Fecha del delito) '''

df['OffenseDate'] = pd.to_datetime(df['OffenseDate'])
```

```python

    '''Creamos una función mediante el uso del método to_datetime para modificar el formato de hora y fecha correctamente dentro de la columna CallDateTime(Fecha y hora de la llamada ) '''

df['CallDateTime'] = pd.to_datetime(df['CallDateTime'])
```

```python

    '''Creamos una función mediante el uso del método to_datetime para modificar el formato de hora correctamente dentro de la columna CallTime, la cual nos presenta el Tiempo de duración de llamada '''

df['CallTime']= pd.to_datetime(df['CallTime'], format='%H:%M:%S', errors='ignore')
```
